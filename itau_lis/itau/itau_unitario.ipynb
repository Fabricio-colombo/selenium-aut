```python
def itau_dic(cpf, matricula, valor_parcela, parcelas_pagas, taxa_origem, saldo_devedor):
    informacoes = {
        'CPF': cpf,
        'Matrícula': matricula,
        'Valor da Parcela': valor_parcela,
        'Parcelas Pagas': parcelas_pagas,
        'Taxa Origem': taxa_origem,
        'Saldo Devedor': saldo_devedor,
    }
    return informacoes
```
